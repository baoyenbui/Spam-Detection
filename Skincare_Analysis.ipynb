# 3. Data Preprocessing

## Các cột chính:
- `author_id`,`product_id`,`rating`,`review_text`,`submission_time`
- `total_feedback_count`,`total_pos_feedback_count`,`total_neg_feedback_count`

## Ý tưởng tiền xử lý : 
* Bộ dữ liệu không có nhãn sẵn nên cần tự gắn nhãn. Để đảm bảo tính khách quan, đánh giá mức độ spam dựa trên nhiều tiêu chí khác nhau bao gồm : graph, behavior, temporal, content và community.
* Mỗi tiêu chí được tính thành một điểm riêng, sau đó kết hợp lại thành một biến tổng hợp.Để tối ưu hóa trọng số của từng tiêu chí trong biến tổng hợp, sử dụng Bayesian optimization.
* Ngoài ra, để tránh mất cân bằng giữa các lớp (spam và non-spam), áp dụng kỹ thuật oversampling như SMOTE.
* Đối với việc phát hiện các pattern bất thường không theo phân phối chính, sử dụng Isolation Forest để bắt các trường hợp nằm ngoài cấu trúc dữ liệu thông thường.

### 1. Graph-Based Detection
- Tạo đồ thị bipartite từ `author_id` và `product_id`
- Tính các chỉ số centrality (degree, betweenness, clustering)
- Phát hiện anomaly nếu centrality > 95th percentile

### 2. Behavioral Analysis
- Thống kê hoạt động user: số review, độ lệch chuẩn, rating pattern
- Phát hiện burst activity và diversity thấp
- So sánh rating với feedback (positive/negative ratio)

### 3. Content Analysis
- Tìm từ khóa spam trong `review_text`
- Phân tích tỷ lệ chữ hoa, dấu câu, URL/email
- So sánh rating với sentiment text
- Kiểm tra consistency với feedback (pos/neg)

### 4. Temporal Patterns
- Phát hiện review đăng lúc 3–5h sáng
- Tìm review trùng ngày từ cùng user
- Khoảng cách giữa các review < 30 phút → nghi vấn

### 5. Community Detection
- Tạo ma trận tương đồng người dùng (cosine similarity)
- Áp dụng Louvain để tìm cộng đồng
- Đánh giá cộng đồng với rating std < 0.5

### 6. Statistical Outliers
- Dùng Isolation Forest để tìm điểm dị biệt
- Dựa trên: độ dài review, rating, feedback
- Phát hiện các pattern bất thường

### 7. Ensemble Optimization
- Tối ưu trọng số 6 components bằng Optuna 
- Label theo nhiều ngưỡng (80%, 85%, 90%)
- Cân bằng dữ liệu với SMOTE sau khi gán nhãn

## Output
- Dataset với các `fraud scores` riêng
- `combined_fraud_score` tối ưu
- Nhãn nhị phân `is_spam` phục vụ mô hình phân loại


```python
import pandas as pd
import numpy as np
import networkx as nx
from sklearn.ensemble import IsolationForest
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import f1_score
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
import optuna
import re
import warnings
warnings.filterwarnings('ignore')

class SpamLabelGenerator:
    def __init__(self):
        self.scaler = StandardScaler()
        self.isolation_forest = IsolationForest(contamination=0.05, random_state=42)
        self.label_encoder = LabelEncoder()
        self.optimal_weights = None
        
    def create_spam_labels(self, reviews_df, products_df):
        print("Creating spam labels...")
        
        graph_fraud_scores = self.detect_graph_fraud(reviews_df, products_df)
        behavioral_fraud_scores = self.detect_behavioral_patterns(reviews_df)
        content_fraud_scores = self.detect_content_fraud(reviews_df)
        temporal_fraud_scores = self.detect_temporal_patterns(reviews_df)
        community_fraud_scores = self.detect_communities(reviews_df)
        statistical_fraud_scores = self.detect_outliers(reviews_df)
        
        component_scores = np.column_stack([
            graph_fraud_scores, behavioral_fraud_scores, content_fraud_scores,
            temporal_fraud_scores, community_fraud_scores, statistical_fraud_scores
        ])
        
        print("Optimizing ensemble weights using Optuna...")
        optimal_weights = self.optimize_weights_optuna(component_scores)
        self.optimal_weights = optimal_weights
        
        combined_scores = self.ensemble_scores_optimized(
            graph_fraud_scores, behavioral_fraud_scores, content_fraud_scores,
            temporal_fraud_scores, community_fraud_scores, statistical_fraud_scores,
            optimal_weights
        )
        
        spam_labels = self.create_balanced_labels(combined_scores)
        
        reviews_df['label'] = spam_labels
        reviews_df['is_spam'] = spam_labels
        
        reviews_df['graph_fraud_score'] = graph_fraud_scores
        reviews_df['behavioral_fraud_score'] = behavioral_fraud_scores
        reviews_df['content_fraud_score'] = content_fraud_scores
        reviews_df['temporal_fraud_score'] = temporal_fraud_scores
        reviews_df['community_fraud_score'] = community_fraud_scores
        reviews_df['statistical_fraud_score'] = statistical_fraud_scores
        reviews_df['combined_fraud_score'] = combined_scores
        
        reviews_df['text'] = reviews_df['review_text']
        
        print(f"Optimal weights: {optimal_weights}")
        print(f"Total samples: {len(reviews_df)}")
        print(f"Spam samples: {sum(spam_labels)} ({sum(spam_labels)/len(spam_labels)*100:.2f}%)")
        print(f"Normal samples: {len(spam_labels) - sum(spam_labels)} ({(len(spam_labels) - sum(spam_labels))/len(spam_labels)*100:.2f}%)")
        
        return reviews_df
    
    def optimize_weights_optuna(self, component_scores, n_trials=50):
        def objective(trial):
            weights = []
            for i in range(6):
                weights.append(trial.suggest_float(f'weight_{i}', 0.05, 0.5))
            
            weights = np.array(weights)
            weights = weights / np.sum(weights)
            
            combined_score = np.dot(component_scores, weights)
            
            high_threshold = np.percentile(combined_score, 90)
            medium_threshold = np.percentile(combined_score, 85)
            low_threshold = np.percentile(combined_score, 80)
            
            y_pred = []
            for score in combined_score:
                if score >= high_threshold:
                    y_pred.append(1)
                elif score >= medium_threshold:
                    y_pred.append(1)
                elif score >= low_threshold:
                    y_pred.append(1)
                else:
                    y_pred.append(0)
            
            y_pred = np.array(y_pred)
            y_true = (combined_score >= np.percentile(combined_score, 80)).astype(int)
            
            if len(np.unique(y_pred)) > 1 and len(np.unique(y_true)) > 1:
                f1 = f1_score(y_true, y_pred)
            else:
                f1 = 0.0
            
            return f1
        
        try:
            study = optuna.create_study(direction='maximize')
            study.optimize(objective, n_trials=n_trials)
            
            optimal_weights = []
            for i in range(6):
                optimal_weights.append(study.best_params[f'weight_{i}'])
            
            optimal_weights = np.array(optimal_weights)
            optimal_weights = optimal_weights / np.sum(optimal_weights)
            
            print(f"Optuna optimization completed. Best F1-score: {study.best_value:.4f}")
            
        except Exception as e:
            print(f"Optuna optimization failed: {e}")
            print("Using default weights...")
            optimal_weights = np.array([0.25, 0.25, 0.20, 0.15, 0.10, 0.05])
        
        return optimal_weights
    
    def detect_graph_fraud(self, reviews_df, products_df):
        G = nx.Graph()
        
        for user_id in reviews_df['author_id'].unique():
            user_reviews = reviews_df[reviews_df['author_id'] == user_id]
            G.add_node(f"user_{user_id}", 
                      node_type='user',
                      review_count=len(user_reviews),
                      avg_rating=user_reviews['rating'].mean())
        
        for product_id in reviews_df['product_id'].unique():
            product_reviews = reviews_df[reviews_df['product_id'] == product_id]
            G.add_node(f"product_{product_id}", 
                      node_type='product',
                      review_count=len(product_reviews),
                      avg_rating=product_reviews['rating'].mean())
        
        for _, row in reviews_df.iterrows():
            G.add_edge(f"user_{row['author_id']}", f"product_{row['product_id']}", 
                      weight=row['rating'], timestamp=row['submission_time'])
        
        fraud_scores = []
        
        degree_centrality = nx.degree_centrality(G)
        betweenness_centrality = nx.betweenness_centrality(G, k=min(1000, len(G.nodes())))
        clustering_coefficient = nx.clustering(G)
        
        for _, row in reviews_df.iterrows():
            user_node = f"user_{row['author_id']}"
            product_node = f"product_{row['product_id']}"
            
            fraud_score = 0
            
            user_degree_cent = degree_centrality.get(user_node, 0)
            if user_degree_cent > np.percentile(list(degree_centrality.values()), 95):
                fraud_score += 3
            
            user_betweenness = betweenness_centrality.get(user_node, 0)
            if user_betweenness > np.percentile(list(betweenness_centrality.values()), 90):
                fraud_score += 2
            
            user_clustering = clustering_coefficient.get(user_node, 0)
            if user_clustering < 0.1 and G.degree(user_node) > 5:
                fraud_score += 2
            
            product_degree = G.degree(product_node)
            if product_degree > np.percentile([G.degree(n) for n in G.nodes() if n.startswith('product_')], 95):
                fraud_score += 1
            
            fraud_scores.append(fraud_score)
        
        max_score = max(fraud_scores) if fraud_scores else 1
        normalized_scores = [score / max_score for score in fraud_scores]
        
        return normalized_scores
    
    def detect_behavioral_patterns(self, reviews_df):
        user_stats = reviews_df.groupby('author_id').agg({
            'rating': ['count', 'mean', 'std', 'min', 'max'],
            'review_text': [lambda x: len(set(x)), lambda x: np.mean([len(text) for text in x])],
            'product_id': 'nunique',
            'submission_time': [lambda x: (pd.to_datetime(x.max()) - pd.to_datetime(x.min())).days,
                               lambda x: len(set(pd.to_datetime(x).dt.date))],
            'total_pos_feedback_count': 'sum',
            'total_neg_feedback_count': 'sum',
            'total_feedback_count': 'sum'
        }).reset_index()
        
        user_stats.columns = ['author_id', 'review_count', 'avg_rating', 'rating_std', 'min_rating', 'max_rating',
                             'unique_reviews', 'avg_review_length', 'unique_products', 'active_days', 'active_dates',
                             'total_positive', 'total_negative', 'total_feedback']
        user_stats = user_stats.fillna(0)
        
        user_stats['review_diversity'] = user_stats['unique_reviews'] / user_stats['review_count']
        user_stats['product_diversity'] = user_stats['unique_products'] / user_stats['review_count']
        user_stats['rating_range'] = user_stats['max_rating'] - user_stats['min_rating']
        user_stats['activity_intensity'] = user_stats['review_count'] / np.maximum(user_stats['active_days'], 1)
        user_stats['burst_activity'] = user_stats['review_count'] / np.maximum(user_stats['active_dates'], 1)
        user_stats['positive_ratio'] = user_stats['total_positive'] / np.maximum(user_stats['total_feedback'], 1)
        user_stats['negative_ratio'] = user_stats['total_negative'] / np.maximum(user_stats['total_feedback'], 1)
        
        fraud_scores = []
        
        for _, row in reviews_df.iterrows():
            user_data = user_stats[user_stats['author_id'] == row['author_id']].iloc[0]
            
            fraud_score = 0
            
            if user_data['review_diversity'] < 0.5 and user_data['review_count'] > 5:
                fraud_score += 3
            
            if user_data['rating_std'] < 0.5 and user_data['review_count'] > 3:
                fraud_score += 2
            
            if user_data['activity_intensity'] > user_stats['activity_intensity'].quantile(0.95):
                fraud_score += 3
            
            if user_data['burst_activity'] > user_stats['burst_activity'].quantile(0.9):
                fraud_score += 2
            
            if user_data['rating_range'] <= 1 and user_data['review_count'] > 2:
                fraud_score += 2
            
            if user_data['avg_review_length'] < 20 or user_data['avg_review_length'] > 1000:
                fraud_score += 1
            
            if user_data['positive_ratio'] < 0.3 or user_data['positive_ratio'] > 0.95:
                fraud_score += 2
            
            if user_data['negative_ratio'] > 0.7:
                fraud_score += 2
            
            fraud_scores.append(fraud_score)
        
        max_score = max(fraud_scores) if fraud_scores else 1
        normalized_scores = [score / max_score for score in fraud_scores]
        
        return normalized_scores
    
    def detect_content_fraud(self, reviews_df):
        fraud_scores = []
        
        promotional_keywords = ['free', 'discount', 'sale', 'offer', 'deal', 'cheap', 'save', 'buy now']
        urgency_keywords = ['urgent', 'limited time', 'act now', 'hurry', 'expires', 'deadline']
        financial_keywords = ['money', 'cash', 'earn', 'profit', 'investment', 'loan', 'credit']
        suspicious_keywords = ['guaranteed', 'risk free', 'no obligation', 'winner', 'congratulations']
        
        for _, row in reviews_df.iterrows():
            text = str(row.get('review_text', '')).lower()
            
            fraud_score = 0
            
            promotional_count = sum(1 for keyword in promotional_keywords if keyword in text)
            urgency_count = sum(1 for keyword in urgency_keywords if keyword in text)
            financial_count = sum(1 for keyword in financial_keywords if keyword in text)
            suspicious_count = sum(1 for keyword in suspicious_keywords if keyword in text)
            
            fraud_score += promotional_count * 1.5
            fraud_score += urgency_count * 2
            fraud_score += financial_count * 2
            fraud_score += suspicious_count * 3
            
            if len(text) > 10:
                caps_ratio = sum(1 for c in text if c.isupper()) / len(text)
                if caps_ratio > 0.3:
                    fraud_score += 3
                
                punct_ratio = sum(1 for c in text if c in '!?') / len(text)
                if punct_ratio > 0.1:
                    fraud_score += 2
                
                if re.search(r'(.{3,})\1{2,}', text):
                    fraud_score += 2
                
                number_ratio = sum(1 for c in text if c.isdigit()) / len(text)
                if number_ratio > 0.2:
                    fraud_score += 1
            
            urls = len(re.findall(r'http[s]?://\S+', text))
            emails = len(re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', text))
            phones = len(re.findall(r'\b\d{3}[-.]?\d{3}[-.]?\d{4}\b', text))
            
            fraud_score += (urls + emails + phones) * 3
            
            words = text.split()
            if len(words) > 0:
                avg_word_length = np.mean([len(word) for word in words])
                if avg_word_length < 3 or avg_word_length > 12:
                    fraud_score += 1
            
            rating = row.get('rating', 3)
            positive_words = ['good', 'great', 'excellent', 'amazing', 'love', 'perfect', 'best']
            negative_words = ['bad', 'terrible', 'awful', 'hate', 'worst', 'horrible', 'disappointing']
            
            positive_count = sum(1 for word in positive_words if word in text)
            negative_count = sum(1 for word in negative_words if word in text)
            
            if rating >= 4 and negative_count > positive_count:
                fraud_score += 2
            elif rating <= 2 and positive_count > negative_count:
                fraud_score += 2
            
            total_feedback = row.get('total_feedback_count', 0)
            total_positive = row.get('total_pos_feedback_count', 0)
            total_negative = row.get('total_neg_feedback_count', 0)
            
            if total_feedback > 0:
                positive_ratio = total_positive / total_feedback
                negative_ratio = total_negative / total_feedback
                
                if positive_ratio < 0.3 or positive_ratio > 0.95:
                    fraud_score += 2
                
                if negative_ratio > 0.7:
                    fraud_score += 2
            
            fraud_scores.append(fraud_score)
        
        max_score = max(fraud_scores) if fraud_scores else 1
        normalized_scores = [score / max_score for score in fraud_scores]
        
        return normalized_scores
    
    def detect_temporal_patterns(self, reviews_df):
        reviews_df['submission_datetime'] = pd.to_datetime(reviews_df['submission_time'])
        reviews_df['hour'] = reviews_df['submission_datetime'].dt.hour
        reviews_df['day_of_week'] = reviews_df['submission_datetime'].dt.dayofweek
        reviews_df['date'] = reviews_df['submission_datetime'].dt.date
        
        fraud_scores = []
        
        for _, row in reviews_df.iterrows():
            fraud_score = 0
            
            if row['hour'] in [3, 4, 5]:
                fraud_score += 2
            
            if row['day_of_week'] in [5, 6]:
                fraud_score += 0.5
            
            user_reviews = reviews_df[reviews_df['author_id'] == row['author_id']]
            
            if len(user_reviews) > 1:
                same_day_reviews = len(user_reviews[user_reviews['date'] == row['date']])
                if same_day_reviews > 3:
                    fraud_score += 2
                
                unusual_hour_ratio = len(user_reviews[user_reviews['hour'].isin([0, 1, 2, 3, 4, 5])]) / len(user_reviews)
                if unusual_hour_ratio > 0.5:
                    fraud_score += 2
                
                time_diffs = user_reviews['submission_datetime'].diff().dt.total_seconds() / 60
                if len(time_diffs) > 1:
                    avg_time_diff = time_diffs.mean()
                    if avg_time_diff < 30:
                        fraud_score += 3
            
            fraud_scores.append(fraud_score)
        
        max_score = max(fraud_scores) if fraud_scores else 1
        normalized_scores = [score / max_score for score in fraud_scores]
        
        return normalized_scores
    
    def detect_communities(self, reviews_df):
        user_product_matrix = reviews_df.pivot_table(
            index='author_id', columns='product_id', values='rating', fill_value=0
        )
        
        from sklearn.metrics.pairwise import cosine_similarity
        user_similarities = cosine_similarity(user_product_matrix)
        
        G = nx.Graph()
        users = user_product_matrix.index.tolist()
        
        for i, user1 in enumerate(users):
            for j, user2 in enumerate(users[i+1:], i+1):
                similarity = user_similarities[i][j]
                if similarity > 0.7:
                    G.add_edge(user1, user2, weight=similarity)
        
        try:
            import community as community_louvain
            partition = community_louvain.best_partition(G)
        except ImportError:
            partition = {}
            for i, component in enumerate(nx.connected_components(G)):
                for node in component:
                    partition[node] = i
        
        community_stats = {}
        for user, comm_id in partition.items():
            if comm_id not in community_stats:
                community_stats[comm_id] = []
            community_stats[comm_id].append(user)
        
        suspicious_communities = set()
        for comm_id, users in community_stats.items():
            if len(users) >= 3:
                comm_reviews = reviews_df[reviews_df['author_id'].isin(users)]
                rating_std = comm_reviews['rating'].std()
                
                if rating_std < 0.5:
                    suspicious_communities.add(comm_id)
        
        fraud_scores = []
        for _, row in reviews_df.iterrows():
            user_id = row['author_id']
            comm_id = partition.get(user_id, -1)
            
            if comm_id in suspicious_communities:
                fraud_scores.append(1.0)
            else:
                fraud_scores.append(0.0)
        
        return fraud_scores
    
    def detect_outliers(self, reviews_df):
        features = []
        
        for _, row in reviews_df.iterrows():
            text = str(row.get('review_text', ''))
            
            feature_vector = [
                len(text),
                len(text.split()),
                row.get('rating', 3),
                text.count('!'),
                text.count('?'),
                sum(1 for c in text if c.isupper()) / max(len(text), 1),
                sum(1 for c in text if c.isdigit()) / max(len(text), 1),
                row.get('total_feedback_count', 0),
                row.get('total_pos_feedback_count', 0),
                row.get('total_neg_feedback_count', 0)
            ]
            features.append(feature_vector)
        
        features = np.array(features)
        outlier_scores = self.isolation_forest.fit_predict(features)
        fraud_scores = [1.0 if score == -1 else 0.0 for score in outlier_scores]
        
        return fraud_scores
    
    def ensemble_scores_optimized(self, *score_arrays, weights=None):
        if weights is None:
            weights = [0.25, 0.25, 0.20, 0.15, 0.10, 0.05]
        
        combined_scores = []
        for i in range(len(score_arrays[0])):
            weighted_score = sum(scores[i] * weight for scores, weight in zip(score_arrays, weights))
            combined_scores.append(weighted_score)
        
        return combined_scores
    
    def create_balanced_labels(self, combined_scores):
        high_threshold = np.percentile(combined_scores, 90)
        medium_threshold = np.percentile(combined_scores, 85)
        low_threshold = np.percentile(combined_scores, 80)
        
        labels = []
        for score in combined_scores:
            if score >= high_threshold:
                labels.append(1)
            elif score >= medium_threshold:
                labels.append(1)
            elif score >= low_threshold:
                labels.append(1)
            else:
                labels.append(0)
        
        return labels

class SMOTEProcessor:
    def __init__(self):
        self.smote = None
        self.smoteen = None
        
    def apply_smote(self, X, y, method='smote'):
        print(f"Applying {method}...")
        
        unique, counts = np.unique(y, return_counts=True)
        print(f"Original distribution: {dict(zip(unique, counts))}")
        
        if len(counts) < 2:
            return X, y
        
        imbalance_ratio = max(counts) / min(counts)
        print(f"Imbalance ratio: {imbalance_ratio:.2f}")
        
        try:
            if method == 'smote':
                k_neighbors = min(5, min(counts) - 1)
                if k_neighbors > 0:
                    self.smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
                    X_resampled, y_resampled = self.smote.fit_resample(X, y)
                else:
                    return X, y
                    
            elif method == 'smoteen':
                k_neighbors = min(3, min(counts) - 1)
                if k_neighbors > 0:
                    self.smoteen = SMOTEENN(random_state=42, 
                                          smote=SMOTE(k_neighbors=k_neighbors, random_state=42))
                    X_resampled, y_resampled = self.smoteen.fit_resample(X, y)
                else:
                    return X, y
            
            unique_resampled, counts_resampled = np.unique(y_resampled, return_counts=True)
            print(f"Resampled distribution: {dict(zip(unique_resampled, counts_resampled))}")
            
            return X_resampled, y_resampled
            
        except Exception as e:
            print(f"SMOTE failed: {e}")
            return X, y

def create_label_column(reviews_df, products_df):
    print("Creating label column for spam detection...")
    
    label_generator = SpamLabelGenerator()
    
    labeled_df = label_generator.create_spam_labels(reviews_df, products_df)
    
    le = LabelEncoder()
    
    categorical_columns = ['author_id', 'product_id']
    for col in categorical_columns:
        if col in labeled_df.columns:
            labeled_df[f'{col}_encoded'] = le.fit_transform(labeled_df[col].astype(str))
    
    labeled_df['label_binary'] = labeled_df['label'].astype(int)
    
    labeled_df['label_confidence'] = labeled_df['combined_fraud_score']
    
    print(f"Label column created successfully!")
    print(f"Label distribution:")
    print(labeled_df['label'].value_counts())
    print(f"Label percentages:")
    print(labeled_df['label'].value_counts(normalize=True) * 100)
    
    return labeled_df

if __name__ == "__main__":
    reviews_df = pd.read_csv("processed_data/memory_optimized_reviews.csv")
    products_df = pd.read_csv("processed_data/memory_optimized_products.csv")
    
    labeled_df = create_label_column(reviews_df, products_df)
    
    feature_columns = ['rating', 'total_feedback_count', 'total_pos_feedback_count', 'total_neg_feedback_count',
                      'graph_fraud_score', 'behavioral_fraud_score', 'content_fraud_score', 
                      'temporal_fraud_score', 'community_fraud_score', 
                      'statistical_fraud_score', 'combined_fraud_score']
    
    X = labeled_df[feature_columns].fillna(0)
    y = labeled_df['label']
    
    smote_processor = SMOTEProcessor()
    X_balanced, y_balanced = smote_processor.apply_smote(X, y, method='smote')
    
    final_df = labeled_df.copy()
    final_df.to_csv("processed_data/spam_detection_with_labels.csv", index=False)
    
    print(f"Final dataset shape: {final_df.shape}")
    print(f"Label column created: 'label'")
    print(f"Additional columns: 'label_binary', 'label_confidence'")
    print(f"Spam rate: {y.mean()*100:.2f}%")
    print(f"Balanced dataset shape: {X_balanced.shape}")
    print(f"Balanced spam rate: {y_balanced.mean()*100:.2f}%")
```

## Fraud Detection: BERT + CNN + Autoencoder + TabNet + Meta-learning

## Ý tưởng xây dựng mô hình
Khi nghĩ đến bài toán xử lý ngôn ngữ tự nhiên, em ngay lập tức nhớ tới BERT vì đã từng đọc một bài nghiên cứu về sentiment analysis sử dụng BERT kết hợp với BiGRU, cho kết quả khá tốt. Từ đó, em muốn vận dụng hướng tiếp cận này vào bài toán hiện tại. Tuy nhiên, thay vì sử dụng BERT đầy đủ, em chọn DistilBERT để tối ưu tốc độ huấn luyện và giảm chi phí tính toán, vì mô hình nhẹ hơn nhưng vẫn giữ được phần lớn chất lượng biểu diễn.

Em cũng có cải tiến ở chỗ đóng băng một phần mô hình BERT (cụ thể là các layer đầu tiên), chỉ fine-tune BiGRU và projection layer ở phía sau. Cách làm này vừa giảm nguy cơ overfitting khi dữ liệu không lớn, vừa tiết kiệm tài nguyên huấn luyện.

Sau đó, trong quá trình phân tích và dựng thử mô hình gợi ý, em nhận thấy rằng dữ liệu khá thưa và rời rạc nên có khả năng tồn tại nhiều pattern ngầm mà BERT khó học được trực tiếp. Vì vậy, em thiết kế thêm một nhánh mô hình sử dụng CNN Autoencoder để trích xuất đặc trưng nén từ các chỉ số kỹ thuật, bao gồm các thống kê về văn bản và các fraud scores đã tính ở bước tiền xử lý. Mục tiêu của nhánh này là học thêm representation sâu hơn, bổ sung cho embedding từ văn bản, đồng thời giảm chiều dữ liệu.

Tiếp theo, em đưa phần đặc trưng bảng chuẩn hóa vào một mô hình TabNet, là một biến thể của MLP có tích hợp attention, giúp tự động chọn lọc đặc trưng quan trọng mà không cần làm thủ công. Em thấy TabNet phù hợp với dữ liệu structured vì khả năng xử lý trực tiếp các dạng bảng phức tạp.

Cuối cùng, em kết hợp toàn bộ các đặc trưng từ DistilBERT + BiGRU, CNN Autoencoder, và TabNet thành một vector hợp nhất, sau đó đưa qua một MLP hai tầng (Decision MLP) để thực hiện decision-level fusion và đưa ra dự đoán cuối cùng.

## Tổng quan
- Multi-modal ensemble learning gồm 4 thành phần chính  
- Kết hợp giữa phân tích văn bản và đặc trưng dạng bảng (tabular features)  
- Decision-level fusion để đưa ra dự đoán cuối cùng  

---

## 1. BERT Encoder
- **Input**: `review_text` (đã tokenize)  
- **Kiến trúc**: DistilBERT (đóng băng) → BiGRU → Projection Layer  
- **Output**: 128-dimensional text embedding  
- **Vai trò**: Hiểu ngữ nghĩa và ngữ cảnh của nội dung review  

---

## 2. CNN Autoencoder
- **Input**: Đặc trưng đã xử lý (text statistics + fraud scores)  
- **Kiến trúc**: 1D CNN encoder–decoder với bottleneck  
- **Output**: 32-dimensional compressed features  
- **Vai trò**: Học representation và phát hiện bất thường (anomaly) trong tabular data  

---

## 3. TabNet Classifier
- **Input**: Scaled tabular features (chuẩn hóa trước)  
- **Kiến trúc**: Attention-based tabular learning (TabNet)  
- **Output**: Xác suất nhị phân (spam / non-spam)  
- **Vai trò**: Mô hình hóa đặc trưng dạng bảng với khả năng chọn lọc đặc trưng tự động  

---

## 4. Decision MLP
- **Input**: Nối (concatenate) tất cả output từ 3 mô hình trên  
- **Kiến trúc**: 2-layer MLP kèm BatchNorm và Dropout  
- **Output**: Dự đoán cuối cùng (binary classification)  
- **Vai trò**: Tổng hợp quyết định từ các nhánh (fusion at decision level)  

---

## Chiến lược Huấn luyện
- **Sequential training**: BERT → CNN-AE → TabNet → Decision MLP  
- **Giới hạn thời gian**: Tối đa 20 phút, có fallback mechanism nếu vượt quá  
- **Class balancing**: Tự động điều chỉnh trọng số lớp dựa trên phân phối dữ liệu  
- **Early stopping**: Áp dụng patience để tránh overfitting  

---

## Feature Engineering
- **Text features**: Độ dài, số dấu câu, tỷ lệ chữ in hoa, tỷ lệ chữ số  
- **Feedback features**: Tỷ lệ feedback tích cực / tiêu cực  
- **Fraud scores**: 6 điểm thành phần từ bước tiền xử lý  
- **Tổng hợp**: 11–17 đặc trưng, tùy vào dữ liệu và mức độ đầy đủ  


```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_score, recall_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from transformers import DistilBertTokenizer, DistilBertModel
from pytorch_tabnet.tab_model import TabNetClassifier
import time
import re
import warnings
warnings.filterwarnings('ignore')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class SpamDetector:
    def __init__(self):
        self.bert_model = None
        self.cnn_autoencoder = None
        self.tabnet_model = None
        self.decision_mlp = None
        self.scaler = StandardScaler()
        self.tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
        self.target_f1 = 0.85
        self.max_time_minutes = 20
        
    def create_bert_encoder(self):
        class BertEncoder(nn.Module):
            def __init__(self):
                super().__init__()
                self.bert = DistilBertModel.from_pretrained('distilbert-base-uncased')
                for param in self.bert.parameters():
                    param.requires_grad = False
                
                self.text_encoder = nn.GRU(768, 128, bidirectional=True, batch_first=True)
                self.text_projection = nn.Linear(256, 128)
                self.dropout = nn.Dropout(0.5)
                
            def forward(self, input_ids, attention_mask):
                with torch.no_grad():
                    bert_output = self.bert(input_ids=input_ids, attention_mask=attention_mask)
                    sequence_output = bert_output.last_hidden_state
                
                gru_output, _ = self.text_encoder(sequence_output)
                pooled = torch.mean(gru_output, dim=1)
                projected = self.text_projection(pooled)
                projected = self.dropout(projected)
                
                return projected
        
        return BertEncoder().to(device)
    
    def create_cnn_autoencoder(self, input_dim):
        class CNNAutoencoder(nn.Module):
            def __init__(self, input_dim):
                super().__init__()
                self.encoder = nn.Sequential(
                    nn.Conv1d(1, 32, kernel_size=3, padding=1),
                    nn.ReLU(),
                    nn.BatchNorm1d(32),
                    nn.MaxPool1d(2),
                    nn.Conv1d(32, 64, kernel_size=3, padding=1),
                    nn.ReLU(),
                    nn.BatchNorm1d(64),
                    nn.AdaptiveAvgPool1d(1)
                )
                
                self.feature_projection = nn.Sequential(
                    nn.Linear(64, 32),
                    nn.ReLU(),
                    nn.Dropout(0.5)
                )
                
                self.decoder = nn.Sequential(
                    nn.Linear(32, 64),
                    nn.ReLU(),
                    nn.Linear(64, input_dim)
                )
                
            def forward(self, x, return_features=False):
                x_reshaped = x.unsqueeze(1)
                conv_features = self.encoder(x_reshaped)
                conv_features = conv_features.squeeze(-1)
                
                encoded_features = self.feature_projection(conv_features)
                
                if return_features:
                    return encoded_features
                
                decoded = self.decoder(encoded_features)
                return encoded_features, decoded
        
        return CNNAutoencoder(input_dim).to(device)
    
    def create_tabnet_classifier(self, input_dim):
        return TabNetClassifier(
            n_d=8, n_a=8,
            n_steps=3,
            gamma=1.2,
            n_independent=1,
            n_shared=1,
            lambda_sparse=1e-2,
            optimizer_fn=torch.optim.Adam,
            optimizer_params=dict(lr=5e-3),
            mask_type='entmax',
            verbose=0
        )
    
    def create_decision_mlp(self, input_dim):
        class DecisionMLP(nn.Module):
            def __init__(self, input_dim):
                super().__init__()
                self.mlp = nn.Sequential(
                    nn.Linear(input_dim, 64),
                    nn.ReLU(),
                    nn.BatchNorm1d(64),
                    nn.Dropout(0.5),
                    nn.Linear(64, 32),
                    nn.ReLU(),
                    nn.Dropout(0.3),
                    nn.Linear(32, 2)
                )
                
            def forward(self, x):
                return self.mlp(x)
        
        return DecisionMLP(input_dim).to(device)
    
    def prepare_features(self, df):
        fraud_feature_cols = [
            'graph_fraud_score', 'behavioral_fraud_score', 'content_fraud_score',
            'temporal_fraud_score', 'community_fraud_score', 'statistical_fraud_score'
        ]
        
        spam_features = []
        for _, row in df.iterrows():
            text = str(row.get('text', row.get('review_text', '')))
            
            feature_vector = [
                len(text),
                len(text.split()),
                text.count('!'),
                text.count('?'),
                sum(1 for c in text if c.isupper()) / max(len(text), 1),
                sum(1 for c in text if c.isdigit()) / max(len(text), 1),
                row.get('rating', 3),
                row.get('total_feedback_count', 0),
                row.get('total_pos_feedback_count', 0),
                row.get('total_neg_feedback_count', 0),
                row.get('sentiment_score', 0)
            ]
            spam_features.append(feature_vector)
        
        spam_features_array = np.array(spam_features)
        
        if all(col in df.columns for col in fraud_feature_cols):
            fraud_features = df[fraud_feature_cols].fillna(0).values
            combined_features = np.hstack([spam_features_array, fraud_features])
        else:
            combined_features = spam_features_array
        
        return combined_features
    
    def train_spam_detection(self, df):
        start_time = time.time()
        
        label_col = 'label' if 'label' in df.columns else ('is_spam' if 'is_spam' in df.columns else 'label_binary')
        text_col = 'text' if 'text' in df.columns else 'review_text'
        
        X_text = df[text_col].astype(str).values
        y_spam = df[label_col].values
        X_features = self.prepare_features(df)
        
        print(f"Training data: {len(X_text)} samples, {sum(y_spam)} spam ({sum(y_spam)/len(y_spam)*100:.1f}%)")
        
        bert_f1 = self.train_bert(X_text, y_spam)
        
        elapsed = (time.time() - start_time) / 60
        if elapsed >= self.max_time_minutes:
            self.create_fallback_decision_mlp()
            return
        
        cnn_loss = self.train_cnn(X_features, y_spam)
        
        elapsed = (time.time() - start_time) / 60
        if elapsed >= self.max_time_minutes:
            self.create_fallback_decision_mlp()
            return
        
        tabnet_f1 = self.train_tabnet(X_features, y_spam)
        
        elapsed = (time.time() - start_time) / 60
        if elapsed >= self.max_time_minutes:
            self.create_fallback_decision_mlp()
            return
        
        self.train_decision(X_text, X_features, y_spam)
    
    def create_fallback_decision_mlp(self):
        self.decision_mlp = nn.Sequential(
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, 2)
        ).to(device)
    
    def train_bert(self, X_text, y, epochs=2):
        X_train, X_val, y_train, y_val = train_test_split(
            X_text, y, test_size=0.2, random_state=42, stratify=y
        )
        
        encoded_train = self.tokenizer.batch_encode_plus(
            X_train.tolist(), add_special_tokens=True, return_attention_mask=True,
            padding=True, max_length=128, truncation=True, return_tensors='pt'
        )
        encoded_val = self.tokenizer.batch_encode_plus(
            X_val.tolist(), add_special_tokens=True, return_attention_mask=True,
            padding=True, max_length=128, truncation=True, return_tensors='pt'
        )
        
        train_dataset = TensorDataset(
            encoded_train['input_ids'],
            encoded_train['attention_mask'],
            torch.LongTensor(y_train)
        )
        val_dataset = TensorDataset(
            encoded_val['input_ids'],
            encoded_val['attention_mask'],
            torch.LongTensor(y_val)
        )
        
        train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=16)
        
        self.bert_model = self.create_bert_encoder()
        
        classifier_head = nn.Sequential(
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, 2)
        ).to(device)
        
        optimizer = torch.optim.AdamW(
            list(self.bert_model.parameters()) + list(classifier_head.parameters()),
            lr=1e-5, weight_decay=0.1
        )
        
        class_counts = np.bincount(y)
        class_weights = torch.FloatTensor([1.0, class_counts[0]/class_counts[1]]).to(device)
        criterion = nn.CrossEntropyLoss(weight=class_weights)
        
        best_f1 = 0
        
        for epoch in range(epochs):
            self.bert_model.train()
            classifier_head.train()
            total_loss = 0
            
            for batch in train_loader:
                input_ids = batch[0].to(device)
                attention_mask = batch[1].to(device)
                labels = batch[2].to(device)
                
                optimizer.zero_grad()
                
                bert_features = self.bert_model(input_ids, attention_mask)
                logits = classifier_head(bert_features)
                loss = criterion(logits, labels)
                
                loss.backward()
                torch.nn.utils.clip_grad_norm_(
                    list(self.bert_model.parameters()) + list(classifier_head.parameters()), 1.0
                )
                optimizer.step()
                
                total_loss += loss.item()
            
            self.bert_model.eval()
            classifier_head.eval()
            all_preds = []
            all_labels = []
            all_probs = []
            
            with torch.no_grad():
                for batch in val_loader:
                    input_ids = batch[0].to(device)
                    attention_mask = batch[1].to(device)
                    labels = batch[2].to(device)
                    
                    bert_features = self.bert_model(input_ids, attention_mask)
                    logits = classifier_head(bert_features)
                    probs = torch.softmax(logits, dim=1)[:, 1]
                    predictions = (probs > 0.4).long()
                    
                    all_preds.extend(predictions.cpu().numpy())
                    all_labels.extend(labels.cpu().numpy())
                    all_probs.extend(probs.cpu().numpy())
            
            val_f1 = f1_score(all_labels, all_preds)
            val_precision = precision_score(all_labels, all_preds, zero_division=0)
            val_recall = recall_score(all_labels, all_preds, zero_division=0)
            val_accuracy = accuracy_score(all_labels, all_preds)
            
            if len(set(all_labels)) > 1:
                val_roc_auc = roc_auc_score(all_labels, all_probs)
            else:
                val_roc_auc = 0.0
                
            val_rmse = np.sqrt(mean_squared_error(all_labels, all_probs))
            
            print(f"BERT Epoch {epoch+1}/{epochs}")
            print(f"  Loss: {total_loss/len(train_loader):.4f}")
            print(f"  Accuracy: {val_accuracy:.4f}")
            print(f"  Precision: {val_precision:.4f}")
            print(f"  Recall: {val_recall:.4f}")
            print(f"  F1-Score: {val_f1:.4f}")
            print(f"  ROC-AUC: {val_roc_auc:.4f}")
            print(f"  RMSE: {val_rmse:.4f}")
            
            if val_f1 > best_f1:
                best_f1 = val_f1
        
        return best_f1
    
    def train_cnn(self, X_features, y, epochs=30):
        X_train, X_val, y_train, y_val = train_test_split(
            X_features, y, test_size=0.2, random_state=42, stratify=y
        )
        
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_val_scaled = self.scaler.transform(X_val)
        
        self.cnn_autoencoder = self.create_cnn_autoencoder(X_train_scaled.shape[1])
        
        X_train_tensor = torch.FloatTensor(X_train_scaled).to(device)
        X_val_tensor = torch.FloatTensor(X_val_scaled).to(device)
        
        optimizer = torch.optim.Adam(self.cnn_autoencoder.parameters(), lr=0.001, weight_decay=0.1)
        criterion = nn.MSELoss()
        
        best_loss = float('inf')
        patience = 3
        patience_counter = 0
        
        for epoch in range(epochs):
            self.cnn_autoencoder.train()
            
            optimizer.zero_grad()
            encoded_features, decoded = self.cnn_autoencoder(X_train_tensor)
            loss = criterion(decoded, X_train_tensor)
            
            loss.backward()
            optimizer.step()
            
            if epoch % 10 == 0:
                self.cnn_autoencoder.eval()
                with torch.no_grad():
                    val_encoded, val_decoded = self.cnn_autoencoder(X_val_tensor)
                    val_loss = criterion(val_decoded, X_val_tensor)
                    
                    print(f"CNN Epoch {epoch+1}: Loss: {val_loss.item():.4f}")
                    
                    if val_loss.item() < best_loss:
                        best_loss = val_loss.item()
                        patience_counter = 0
                    else:
                        patience_counter += 1
                        if patience_counter >= patience:
                            break
        
        return best_loss
    
    def train_tabnet(self, X_features, y):
        X_train, X_test, y_train, y_test = train_test_split(
            X_features, y, test_size=0.3, random_state=42, stratify=y
        )
        
        X_train_scaled = self.scaler.transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)
        
        self.tabnet_model = self.create_tabnet_classifier(X_train_scaled.shape[1])
        
        self.tabnet_model.fit(
            X_train_scaled, y_train,
            eval_set=[(X_test_scaled, y_test)],
            max_epochs=30,
            patience=5,
            batch_size=512,
            virtual_batch_size=256,
            num_workers=0,
            drop_last=False
        )
        
        tabnet_pred = self.tabnet_model.predict(X_test_scaled)
        tabnet_proba = self.tabnet_model.predict_proba(X_test_scaled)[:, 1]
        
        val_f1 = f1_score(y_test, tabnet_pred)
        val_precision = precision_score(y_test, tabnet_pred, zero_division=0)
        val_recall = recall_score(y_test, tabnet_pred, zero_division=0)
        val_accuracy = accuracy_score(y_test, tabnet_pred)
        
        if len(set(y_test)) > 1:
            val_roc_auc = roc_auc_score(y_test, tabnet_proba)
        else:
            val_roc_auc = 0.0
            
        val_rmse = np.sqrt(mean_squared_error(y_test, tabnet_proba))
        
        print("TabNet Results:")
        print(f"  Accuracy: {val_accuracy:.4f}")
        print(f"  Precision: {val_precision:.4f}")
        print(f"  Recall: {val_recall:.4f}")
        print(f"  F1-Score: {val_f1:.4f}")
        print(f"  ROC-AUC: {val_roc_auc:.4f}")
        print(f"  RMSE: {val_rmse:.4f}")
        
        return val_f1
    
    def train_decision(self, X_text, X_features, y):
        bert_features = self.extract_bert_features(X_text)
        cnn_features = self.extract_cnn_features(X_features)
        tabnet_features = self.extract_tabnet_features(X_features)
        
        combined_features = np.hstack([
            bert_features,
            cnn_features,
            tabnet_features
        ])
        
        X_train, X_test, y_train, y_test = train_test_split(
            combined_features, y, test_size=0.2, random_state=42, stratify=y
        )
        
        input_dim = combined_features.shape[1]
        self.decision_mlp = self.create_decision_mlp(input_dim)
        
        train_dataset = TensorDataset(
            torch.FloatTensor(X_train),
            torch.LongTensor(y_train)
        )
        test_dataset = TensorDataset(
            torch.FloatTensor(X_test),
            torch.LongTensor(y_test)
        )
        
        train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=64)
        
        optimizer = torch.optim.Adam(self.decision_mlp.parameters(), lr=0.001, weight_decay=0.1)
        
        class_counts = np.bincount(y)
        class_weights = torch.FloatTensor([1.0, class_counts[0]/class_counts[1]]).to(device)
        criterion = nn.CrossEntropyLoss(weight=class_weights)
        
        best_f1 = 0
        patience = 3
        patience_counter = 0
        
        for epoch in range(30):
            self.decision_mlp.train()
            total_loss = 0
            
            for batch_x, batch_y in train_loader:
                batch_x = batch_x.to(device)
                batch_y = batch_y.to(device)
                
                optimizer.zero_grad()
                outputs = self.decision_mlp(batch_x)
                loss = criterion(outputs, batch_y)
                loss.backward()
                optimizer.step()
                
                total_loss += loss.item()
            
            if epoch % 10 == 0:
                self.decision_mlp.eval()
                all_preds = []
                all_labels = []
                all_probs = []
                
                with torch.no_grad():
                    for batch_x, batch_y in test_loader:
                        batch_x = batch_x.to(device)
                        batch_y = batch_y.to(device)
                        
                        outputs = self.decision_mlp(batch_x)
                        predictions = torch.argmax(outputs, dim=1)
                        probs = torch.softmax(outputs, dim=1)[:, 1]
                        
                        all_preds.extend(predictions.cpu().numpy())
                        all_labels.extend(batch_y.cpu().numpy())
                        all_probs.extend(probs.cpu().numpy())
                
                current_f1 = f1_score(all_labels, all_preds)
                current_accuracy = accuracy_score(all_labels, all_preds)
                current_precision = precision_score(all_labels, all_preds, zero_division=0)
                current_recall = recall_score(all_labels, all_preds, zero_division=0)
                
                if len(set(all_labels)) > 1:
                    current_roc_auc = roc_auc_score(all_labels, all_probs)
                else:
                    current_roc_auc = 0.0
                    
                current_rmse = np.sqrt(mean_squared_error(all_labels, all_probs))
                
                print(f"Decision MLP Epoch {epoch+1}/30")
                print(f"  Loss: {total_loss/len(train_loader):.4f}")
                print(f"  Accuracy: {current_accuracy:.4f}")
                print(f"  Precision: {current_precision:.4f}")
                print(f"  Recall: {current_recall:.4f}")
                print(f"  F1-Score: {current_f1:.4f}")
                print(f"  ROC-AUC: {current_roc_auc:.4f}")
                print(f"  RMSE: {current_rmse:.4f}")
                
                if current_f1 > best_f1:
                    best_f1 = current_f1
                    patience_counter = 0
                else:
                    patience_counter += 1
                    if patience_counter >= patience:
                        break
        
        print("\nFinal Results:")
        print(f"  Accuracy: {current_accuracy:.4f}")
        print(f"  Precision: {current_precision:.4f}")
        print(f"  Recall: {current_recall:.4f}")
        print(f"  F1-Score: {current_f1:.4f}")
        print(f"  ROC-AUC: {current_roc_auc:.4f}")
        print(f"  RMSE: {current_rmse:.4f}")
    
    def extract_bert_features(self, X_text):
        encoded = self.tokenizer.batch_encode_plus(
            X_text.tolist(), add_special_tokens=True, return_attention_mask=True,
            padding=True, max_length=128, truncation=True, return_tensors='pt'
        )
        
        dataset = TensorDataset(encoded['input_ids'], encoded['attention_mask'])
        dataloader = DataLoader(dataset, batch_size=16)
        
        bert_features = []
        
        self.bert_model.eval()
        with torch.no_grad():
            for batch in dataloader:
                input_ids = batch[0].to(device)
                attention_mask = batch[1].to(device)
                
                features = self.bert_model(input_ids, attention_mask)
                bert_features.append(features.cpu().numpy())
        
        return np.vstack(bert_features)
    
    def extract_cnn_features(self, X_features):
        X_features_scaled = self.scaler.transform(X_features)
        X_tensor = torch.FloatTensor(X_features_scaled).to(device)
        
        self.cnn_autoencoder.eval()
        with torch.no_grad():
            features = self.cnn_autoencoder(X_tensor, return_features=True)
        
        return features.cpu().numpy()
    
    def extract_tabnet_features(self, X_features):
        X_features_scaled = self.scaler.transform(X_features)
        tabnet_proba = self.tabnet_model.predict_proba(X_features_scaled)
        return tabnet_proba
    
    def predict_spam(self, df, threshold=0.4):
        text_col = 'text' if 'text' in df.columns else 'review_text'
        X_text = df[text_col].astype(str).values
        X_features = self.prepare_features(df)
        
        if self.decision_mlp is None:
            bert_features = self.extract_bert_features(X_text)
            
            with torch.no_grad():
                features_tensor = torch.FloatTensor(bert_features).to(device)
                predictions = torch.sigmoid(torch.mean(features_tensor, dim=1))
            
            return (predictions.cpu().numpy() > threshold).astype(int)
        
        bert_features = self.extract_bert_features(X_text)
        cnn_features = self.extract_cnn_features(X_features)
        tabnet_features = self.extract_tabnet_features(X_features)
        
        combined_features = np.hstack([
            bert_features,
            cnn_features,
            tabnet_features
        ])
        
        self.decision_mlp.eval()
        with torch.no_grad():
            features_tensor = torch.FloatTensor(combined_features).to(device)
            outputs = self.decision_mlp(features_tensor)
            predictions = torch.softmax(outputs, dim=1)[:, 1]
        
        return (predictions.cpu().numpy() > threshold).astype(int)

if __name__ == "__main__":
    try:
        reviews_df = pd.read_csv("processed_data/spam_detection_with_labels.csv")
        print(f"Loaded preprocessed data: {reviews_df.shape}")
        
        if 'label' in reviews_df.columns:
            print("Label distribution:", reviews_df['label'].value_counts())
        
    except FileNotFoundError:
        alternative_paths = [
            "spam_detection_with_labels.csv",
            "processed_data/memory_optimized_reviews.csv",
            "memory_optimized_reviews.csv"
        ]
        
        reviews_df = None
        for path in alternative_paths:
            try:
                reviews_df = pd.read_csv(path)
                print(f"Loaded data from: {path}")
                break
            except FileNotFoundError:
                continue
        
        if reviews_df is None:
            print("No data file found.")
            exit()
        
        if 'label' not in reviews_df.columns:
            try:
                from SpamLabelGenerator import create_label_column
                products_df = pd.read_csv("processed_data/memory_optimized_products.csv")
                reviews_df = create_label_column(reviews_df, products_df)
            except Exception as e:
                print(f"Error in label generation: {e}")
                exit()
    
    spam_detector = SpamDetector()
    spam_detector.train_spam_detection(reviews_df)
    
    sample_data = reviews_df.head(10)
    predictions = spam_detector.predict_spam(sample_data)
    
    label_col = 'label' if 'label' in sample_data.columns else 'is_spam'
    
    print(f"\nSample predictions: {predictions}")
    print(f"Actual labels: {sample_data[label_col].values}")
    
    print(f"\nDataset shape: {reviews_df.shape}")
    print(f"Spam rate: {reviews_df[label_col].mean()*100:.2f}%")
```
